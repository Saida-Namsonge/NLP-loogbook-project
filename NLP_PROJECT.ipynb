```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

# Sample text
text = "This is a sample sentence showing stopword removal."

# Get English stopwords and tokenize
stop_words = set(stopwords.words('english'))
tokens = word_tokenize(text.lower())

# Remove stopwords
filtered_tokens = [word for word in tokens if word not in stop_words]

print("Original:", tokens)
print("Filtered:", filtered_tokens)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
text = "NLTK is a powerful library for natural language processing."
words = word_tokenize(text)

pos_tags = pos_tag(words)
print("Original Text:")
print(text)

print("\nPoS Tagging Result:")
for word, pos_tag in pos_tags:
    print(f"{word}: {pos_tag}")
```

```python
from collections import defaultdict
from nltk.tokenize import word_tokenize
import nltk


nltk.download('punkt')


text = "I love studying Natural Language Processing"


tokens = word_tokenize(text)

# Build prediction dictionary
prediction = defaultdict(list)

for i in range(len(tokens) - 1):
    prediction[tokens[i]].append(tokens[i + 1])

# Display predictions
for word in prediction:
    print(word, "->", prediction[word])
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag


nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')


text = "Natural Language Processing is transforming the world of Artificial Intelligence. It is a key area of AI."


sentences = sent_tokenize(text)




for i, sent in enumerate(sentences, start=1):
    print(f"\nSentence {i}: {sent}")


    tokens = word_tokenize(sent)


    tagged = pos_tag(tokens)

    print("\nWord-by-word analysis:")


    for word, tag in tagged:
        print(f"{word:15} -> {tag}")

```

```python
import nltk
import random
from nltk.corpus import treebank
from nltk.tag import hmm

nltk.download('treebank')


tagged_sentences = treebank.tagged_sents()


train_data = tagged_sentences[:3000]
test_data = tagged_sentences[3000:]


trainer = hmm.HiddenMarkovModelTrainer()
hmm_model = trainer.train(train_data)


sentence = "The president visited Nairobi".split()


print("Sequence Labeling Output:")
print(hmm_model.tag(sentence))
```

```python
print(hmm_model.evaluate(test_data))
```

```python
import nltk
from nltk.corpus import treebank
from nltk.tag import hmm
nltk.download('treebank')
tagged_sentences = treebank.tagged_sents()
print("Total sentences in dataset:", len(tagged_sentences))
train_data = tagged_sentences[:3000]
test_data = tagged_sentences[3000:]
print("Training sentences:", len(train_data))
print("Testing sentences:", len(test_data))
trainer = hmm.HiddenMarkovModelTrainer()
hmm_model = trainer.train(train_data)
sentence = "The president visited Nairobi".split()
tagged_output = hmm_model.tag(sentence)
for word, tag in tagged_output:
    print(f"{word:15} -> {tag}")
sentence2 = "Natural Language Processing is powerful".split()
print(hmm_model.tag(sentence2))
accuracy = hmm_model.evaluate(test_data)
print("Accuracy:", accuracy)
print(hmm_model.probability(train_data[0]))
```
